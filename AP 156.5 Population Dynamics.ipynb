\begin{equation}
x_{n+1} = 4rx_n(1-x_n) = f(x_n), n = 0,1,2,3,4,5\\
x = 4r x_n - 4rx_n^2\\
dx = 4r - 8rx_n
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
def f(xn):
    return 4*r*xn*(1-xn)
```

```python
xo = 0.65
r = 0.87
```

```python
xn = []
for n in range(0,45):
    xn.append(xo)
    xo = f(xo)
```

```python
print(xn[5],xn[9],xn[13],xn[17])
```

```python
plt.plot(range(0,45),xn,'.')
plt.plot(range(0,45),xn,'-')
plt.ylabel('x(n)')
plt.xlabel('n')
plt.title("Fig. 3.1. Iteration starting from xo = 0.65 using the parameter r = 0.87. ")
plt.show()
```

```python
x = np.linspace(0,1,150)
for n in range(1,5):
    x = [f(xi) for xi in x]
    if n==1 or n==2 or n==4:
        plt.plot(np.linspace(0,1,150),x, label= str(n)+"-fold iterations")
plt.plot(np.linspace(0,1,150),np.linspace(0,1,150))
plt.legend()
plt.ylabel(r'$f^{n}$(x)')
plt.xlabel('x')
plt.title("Fig 3.2 The function f(x) and its 2-fold and 4-fold iteration")
plt.show()
```
