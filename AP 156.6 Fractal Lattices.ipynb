```python
import numpy as np
import matplotlib.pyplot as plt
import random as rn
rn.seed(5)
```

```python
def qn(qo,r):
    return [(p[r][0]+qo[0])/2,(p[r][1]+qo[1])/2]
```

```python
def m(p1,p2):
    return [(p1[0]+p2[0])/2,(p1[1]+p2[1])/2]
```

```python
p = [[rn.randint(0, 9),rn.randint(0, 9)],
     [rn.randint(0, 9),rn.randint(0, 9)],
     [rn.randint(0, 9),rn.randint(0, 9)]]
q = [2,3]
p=np.array(p)
```

```python
for n in range(0,10000):
    plt.scatter(q[0],q[1],s=0.5)
    r = (rn.randint(0, 2))
    q = qn(q,r)
plt.show()
```

```python
def tri(p,c=0,ctr=4):
    if c<ctr:
        c+=1
        for i in range(len(p)):
            points = [p[i]]
            for j in range(len(p)):
                mid = m(p[i],p[j])
                points.append(mid)
                zipp = list(zip(p[i],mid))
                plt.plot(zipp[0],zipp[1],'k')
            tri(points,c)

tri(p)
plt.show()
```
